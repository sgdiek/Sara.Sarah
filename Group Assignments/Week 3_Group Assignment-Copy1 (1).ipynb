```python
import folium
```

```python
# Create a folium Choropleth map of the percent of households with at least one child by Census tract in SF County
# The map does not output for some reason, but when saved and opened as an HTML it becomes visible and interactive
m = folium.Map(location =[37.7749,-122.4194],
              zoom_start=9,
              tiles='OpenStreetMap')
folium.Choropleth(
                    geo_data=tracts_HH,
                    data=tracts_HH,
                    key_on='feature.properties.FIPS',
                    columns=['FIPS','PCT_HH_One_or_More_Under_18'],
                    fill_color='BuPu',
                    line_weight=0.1,
                    fill_opacity=0.8,
                    line_opacity=0.2,
                    legend_name='Percent of Households With One or More Children (2014-2018 ACS 5-Year Estimates').add_to(m)
```

```python
# Save map as an HTML
m.save('SF_PCT_HH_Atleast_One_Child.html')
```
